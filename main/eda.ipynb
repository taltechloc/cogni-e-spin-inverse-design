```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/PVA-Ziabari-2009.csv")
df.drop(columns=["diameter_stdev"], inplace=True)
df.head()
```

```python
data = df['diameter']
mean_diam = np.mean(data)
std_diam = np.std(data)
median_diam = np.median(data)
n = len(data)

plt.rcParams['font.family'] = 'serif'

fig, ax = plt.subplots(figsize=(3.5, 2.8), dpi=300)

sns.histplot(data, kde=True, bins=12, color='steelblue', edgecolor='black', linewidth=0.3, alpha=0.7)

plt.xlabel("Diameter (nm)", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.title("Nanofiber Diameter Distribution", fontsize=10) 
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

stats_text = f'n = {n}\nMean = {mean_diam:.1f} nm\nSD = {std_diam:.1f} nm\nMedian = {median_diam:.1f} nm'
ax.text(0.95, 0.95, stats_text, transform=ax.transAxes, fontsize=7,
        verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8, pad=0.3, linewidth=0.5))

plt.tight_layout()

plt.savefig('figure_1.pdf', bbox_inches='tight')
plt.savefig('figure_1.png', dpi=300, bbox_inches='tight')

plt.show()
```
